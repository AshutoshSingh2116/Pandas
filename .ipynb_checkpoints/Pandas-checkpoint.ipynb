# MAKING CHANGES TO THE DATA

## (a): Adding a particular column.

```python
df=pd.read_csv('Pokemon_datasets.csv')        #Reading data from csv file.
print(df)             #Prints the data present in df(data_frame) variable from csv file.
```

```python
df['Total']= df['HP'] + df['Attack'] + df['Defense']   #This creates a column as 'Total' and adds the values of each row 
#and stores it in the 'total' column.
df.head(3)    #This prints the first three elements of the dataset.
```

## (b): Removing a particular column 

```python
df=df.drop(columns=['Total'])
df.head(3)
```

```python
df['Total']=df.iloc[:,4:10].sum(axis=1)  #This adds a column 'Total' at axis 1(left to right) from column number 4 to n-1(4 to 10).
#Here, [:,4:10] means- 4:10 addresses columns from the dataset and the part between opening of square brackets to comma addresses the rows.
#Since, we can see that rows are not defined, which means it would take all the rows.
df.head(3)       #This prints the first three rows.
```

```python
#Have a look into the "Name" column, and let us say that we want to point out all the names which contain "Mega" in them!
df.loc[df['Name'].str.contains('Mega')]
```

```python
#Now, let us say, we want all other names first in the table, except which contains 'Mega' in them.
df.loc[~df['Name'].str.contains('Mega')]
```

```python
#Now, lets say, we want all the other names except the names which contain 'Mega' in them.
df.loc[~df['Name'].str.contains('Mega')]
```

```python
#Wanted to check if the 'Type 1' was either 'Grass' of 'Fire':
import re             #re: Regular Expressions: A powerful tool used to search, match and manipulate strings based on data patterns.
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
#This line filters your DataFrame df and returns only those rows where the column Type 1 contains either "Fire" or "Grass".
```

```python
#Here, we add 'flags=re.I'  bcoz it ignores the case, in which the data is written.
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
```

```python
df.loc[df['Name'].str.contains('pikachu', flags=re.I, regex=True)]
#Seached a particular data in the name column, which later showed me the row in which it is present.
```

```python
#If we want all the names which begin with the letter "Pi" in the 'Names' column.
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
#Here, ^pi[a-z]*: '^' means starting letter should be 'Pi' and later it can be followed by any alphabet from 'a-z' and '*' means it 
#can be a ZERO or more.
```

```python
#If we don't need the starting letter as 'Pi', we will just erase '^' which is placed before pi.
#eg: df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
df.loc[df['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)]
#So, this will give names which contain 'Pi', anywhere in them.
```

# AGGREGATE STATISTICS(Groupby)

```python
#Just loading the saved csv file.
df = pd.read_csv('Pokemon_datasets.csv')
df
```

```python
df.dtypes
```

## (a): mean()

```python
#This gives the mean of all the columns corresponding to Rows of Type 1.
df = df.groupby(['Type 1']).mean()
df
```

```python
#This gives the mean of all the columns corresponding to Rows of Type 1, also it sorts the values of column named HP
# in descending order.
new_df=df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)
new_df
```

## (b): sum()

```python
#This groups the DataFrame by the values in the 'Type 1' column.
#Then, it sums all the numeric columns (like HP, Attack, etc.) for each unique 'Type 1'.
#Non-numeric columns are ignored during summation.
df.groupby(['Type 1']).sum()
```

## (c): count()

```python
#This counts the similar rows of Type 1 columns and displays it.
df.groupby(['Type 1']).count()
```

```python
df['Count']=1
df
df.groupby(['Type 1']).count()['Count']
#df['Type 1'].value_counts()
```

```python
#The line df.groupby(['Type 1', 'Type 2']).count()['Count'] groups the data by both Type 1 and Type 2,
#then counts how many rows are in each group using the 'Count' column (which was earlier set to 1 for every row). 
#This gives you the number of Pokémon for each unique type combination.
df.groupby(['Type 1', 'Type 2']).count()['Count']
```

# Working with Large amounts of data

```python
for df in pd.read_csv('Modified_CSV.csv', chunksize=5):
    print ("CHUNK DF")
    print(df)
#This divides the data frame into chunks, where each chunks consists of size of 5 rows.
```

```python
#Adding another dataset to an older dataset and now, we can carry on with other operations on these two combined data frames.
new_df= pd.DataFrame(columns=df.columns)

for df in pd.read_csv('Modified_CSV.csv', chunksize=5):
    results=df.groupby(['Type 1']).count()

    new_df=pd.concat([new_df, results])
    print(new_df)
```
